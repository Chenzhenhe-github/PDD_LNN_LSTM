# Obtain age and gender information for the training set or test set from statistical files

```python
import os
import numpy as np
import pandas as pd
from natsort import natsorted

# Dataset path
path = 'D:\\python test\\project83\\buchong_beichang\\'

# Read file names
total_name = pd.read_excel(path+'total.xlsx', engine='openpyxl') 

# Split file names (training and prediction)
train_list = natsorted(os.listdir(path+'training'))
train_name = [j.split('.')[0] for j in train_list]
train_df = pd.DataFrame(train_name,columns=['name'])

prediction_list = natsorted(os.listdir(path+'prediction'))
prediction_name = [j.split('.')[0] for j in prediction_list]
prediction_df = pd.DataFrame(prediction_name,columns=['name'])

# Take union of file names
prediction_stas = pd.merge(total_name,prediction_df,on='name',how='inner')
prediction_stas.to_excel(path+'prediction_store\\'+'agesex.xlsx',index=False)
train_stas = pd.merge(total_name,train_df,on='name',how='inner')
train_stas.to_excel(path+'training_store\\'+'agesex.xlsx',index=False)
# Merge is tricky: find intersection based on right dataframe's column name matching left dataframe
```

## Random splitting of dataset

```python
# Generate folders for the data to be processed
import os
import shutil
import numpy as np
import pandas as pd
from natsort import natsorted
from sklearn.model_selection import train_test_split

# Initial parameter settings
randoms = 88
group_num = [114,103,209]
path = 'D:\\python test\\project83\\final\\pos5\\'

# Define function to create directory if not exists
def create_directory_if_not_exists(dir_path):
    if not os.path.exists(dir_path):
        os.makedirs(dir_path)
        print(f"目录 {dir_path} 已创建")
    else:
        print(f"目录 {dir_path} 已存在")      
train_destination = str(create_directory_if_not_exists(path+'training\\'))
prediction_destination = str(create_directory_if_not_exists(path+'prediction\\'))

# Dataset split
def file_transfer(file_name,path_source,path_destination):
    for name in file_name:
        source_file = path_source + name
        destination_file = path_destination + name
        shutil.copy2(source_file, destination_file)

        # Random split of filenames
name_list = natsorted(os.listdir(path+'total'))

# Dataset split
train_label = [((np.ones(value))*index).tolist() if index > 0 else np.zeros(value).tolist() for index,value in enumerate(group_num)]
train_target = np.array(sum(train_label,[]))

# Split indices instead of splitting y_test
train_name,test_name,train_target,test_target = train_test_split(name_list,train_target,train_size=0.8,random_state=randoms)

# File transfer
file_transfer(train_name,path+'total\\',path+'training\\')
file_transfer(test_name,path+'total\\',path+'prediction\\')
```
